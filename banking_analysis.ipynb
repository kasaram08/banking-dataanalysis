```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'name' : ['kiran', 'kumar', 'reddy'],
    'account' : ['savings', 'current', 'demat'],
    'balance' : [100, 200, 567],
    'age' : [23 ,66 ,35]
}
df = pd.DataFrame(data)
print(df)
df = pd.DataFrame(data)
df.plot(kind='bar', x='name', y='balance', color='blue')
plt.title('Bank balance of each person')
plt.xlabel('name')
plt.ylabel('balance')
df.plot (kind='pie',y='balance',label=df['name'], autopct='%1.1f%%')
plt.title('Balance Distrubution')
data = {
    'month':['jan', 'feb', 'mar', 'apr', 'may'],
    'balance':[600,200,250,400,530]
} 
df=pd.DataFrame(data)
df.plot(kind='line', x='month',y='balance', color='green', marker='o')
plt.title('Balance Change Over Months')
plt.xlabel('Month')
plt.ylabel('Balance')
plt.show()
```
